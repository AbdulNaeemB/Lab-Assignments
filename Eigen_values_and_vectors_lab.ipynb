```python
import numpy as np
import pandas as pd
import scipy.linalg as la
```

```python
A = np.array([[1,-1],[1,-2]])
result = la.eig(A)
print("Eigen values\n")
print(result[0])
print("Eigen vectors\n")
print(result[1])
```

```python
eigen_values = result[0]
eigen_vectors = result[1]
print("Real eigen values")
print(eigen_values.real)
```

```python
eigvalues , eigvectors = la.eig(A)
print(eigvalues)
print(eigvectors)
print(eigvalues.real)
```

```python
lambda1 = eigvalues[1].real
v1 =  eigvectors[:,1].reshape(2,1)
```

```python
v1 * lambda1
```

```python
n = 4
P = np.random.randint(0,10,(n,n))
print(P)
```

```python
D = np.diag((3,1))
print(D)
```

```python
S = P @ P.T
print(S)
```

```python
evals , evecs = la.eig(S)
print(evals)
print(evecs)
```

```python
v1 = evecs[:,0]
print(v1)
v2 = evecs[:,1]
print(v2)
```

```python
v1 @ v2
```

```python
D = np.diag((3,2,1,4))
print(D)
```

```python
M = P @ D @ la.inv(P)
evals1 , evecs1 = la.eig(M)
print(evals1)
print(evecs1)
```

```python
Pinv = la.inv(P)
k=20
```

```python
%%timeit
result = M.copy()
for _ in range(1,k):
  result = result + M
```

```python
%%timeit
P @ D**k @ Pinv
```
